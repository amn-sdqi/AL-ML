```python
data_melt=pd.melt(df, id_vars=['Date', 'Drug_Name','Parameter'], var_name='Time', value_name='Readings')
data_melt
```

```python
data_tidy=data_melt.pivot(index=['Date','Time','Drug_Name'],columns='Parameter', values="Readings").reset_index()
data_tidy
```

```python
data_tidy.columns.name=None
data_tidy
```

```python
pd.pivot_table(data_tidy, index='Drug_Name', columns='Date', values=['Pressure','Temperature'])
```
