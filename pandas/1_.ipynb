```python
movies= pd.read_csv('movies.csv', index_col=0)
movies
```

```python
directors=pd.read_csv('directors.csv', index_col=0)
directors
```

```python
df=movies.merge(directors, left_on='director_id', right_on='id')
df
```

```python
df.drop(['id_y','director_id'], axis=1,inplace=True)
df.rename({"id_x":'ID'}, axis=1, inplace=True)
```

```python
df[:1]
```

```python
def profit(df):
    return df['budget']-df['revenue']
```

```python
df['Profit']=df[['budget','revenue']].apply(profit, axis=1)
```

```python
df
```

```python
df['vote_average'].apply(round)
```

```python
df.groupby(['director_name']).get_group('Woody Allen')
```

```python
df.groupby(['director_name']).aggregate({'vote_average':['min','max','mean']})
```
