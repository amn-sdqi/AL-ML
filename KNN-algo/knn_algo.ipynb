```python
score_list=[]
for i in range(1,51):
    knn = KNeighborsClassifier(n_neighbors=i ,metric='euclidean')
    knn.fit(x_sm,y_sm)
    score=round(knn.score(X_val,Y_val)*100, 2)
    score_list.append(score)
    
```
